```python
sheet_df = pd.DataFrame()
```

```python
sheet_df["S.No"] = df1["S.No"]
sheet_df["Program"] = df1["Program"]
sheet_df["Semester"] = df1["Semester"]
sheet_df["Course Code"] = df1["Course Code"]
sheet_df["Course Name"] = df1["Course Name"]
sheet_df["No. of Students"] = df1["No. of Students"]
```

```python
sheet_df
```

```python
import numpy as np
import pandas as pd

# Splitting 'L-T-P' column and converting to numeric
sheet_df[['L', 'T', 'P']] = df1['L-T-P'].str.split('-', expand=True)
sheet_df[['L', 'T', 'P']] = sheet_df[['L', 'T', 'P']].apply(pd.to_numeric)

# Filtering non-zero lectures and tutorials
sheet_df = sheet_df[(sheet_df['L'] != 0) | (sheet_df['T'] != 0)]

# Define start and end dates
start_date = '2024-03-11'
end_date = '2024-04-28'

# Create date range and select weekdays
date_range = pd.date_range(start=start_date, end=end_date)
weekdays_range = date_range[date_range.dayofweek < 5]

# Define time slots based on term
term = "end-term"  # or "end-term"
if term == "mid-term":
    time_slots = [
        '9:00 - 10:30', 
        '11:30 - 13:00', 
        '14:00 - 15:30'
    ]
elif term == "end-term":
    time_slots = [
        '9:00 - 12:30', 
        '13:00 - 15:30'
    ]

date_index = 0
time_slot_index = 0 
last_date = {} 
prev_program_semester = None  

for i, row in sheet_df.iterrows():
    program_semester = (row["Program"], row["Semester"]) 
    
    if pd.isna(row["Course Name"]):  
        sheet_df.at[i, 'Date'] = np.nan  
        date_index = 0  
    else:
        if program_semester != prev_program_semester:
            date_index = 0  
            prev_program_semester = program_semester
            time_slot_index = 0 
        
        if program_semester in last_date:
            last_assigned_date = last_date[program_semester] + pd.Timedelta(days=2) 
        else:
            last_assigned_date = pd.Timestamp(weekdays_range[0])  
        
        while True:
            if date_index >= len(weekdays_range):
                # Reset date index if all dates are used
                date_index = 0
                
            if weekdays_range[date_index].dayofweek < 5:
                current_date = pd.Timestamp(weekdays_range[date_index].date())  
                if current_date >= last_assigned_date:
                    sheet_df.at[i, 'Date'] = current_date  
                    last_date[program_semester] = current_date 
                    
                    sheet_df.at[i, 'Time Slot'] = time_slots[time_slot_index]
                    
                    time_slot_index = (time_slot_index + 1) % len(time_slots)
                    
                    date_index += 1  
                    break
                else:
                    date_index += 1  
            else:
                date_index += 1  

```

```python
sheet_df['Year'] = ''

# Define conditions and assign values to the "Year" column based on the "Semester" values
sheet_df.loc[sheet_df['Semester'].isin(['I', 'II']), 'Year'] = 1
sheet_df.loc[sheet_df['Semester'].isin(['III', 'IV']), 'Year'] = 2
sheet_df.loc[sheet_df['Semester'].isin(['V', 'VI']), 'Year'] = 3
sheet_df.loc[sheet_df['Semester'].isin(['VII', 'VIII']), 'Year'] = 4
sheet_df.loc[sheet_df['Semester'].isin(['IX', 'X']), 'Year'] = 5
```
