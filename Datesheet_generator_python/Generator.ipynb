```python
import pandas as pd
import xlsxwriter
```

```python
excel_file = 'BBA & MBA PROGRAM STRUCTURE EVEN SEMESTER 2024 (2).xlsx'
```

```python

df = pd.read_excel(excel_file)
```

```python
df
```

```python
df2 = pd.DataFrame()
```

```python
df2["S.no."] = df["S No"]
```

```python
df2["SUBJECT CODES"] = df["SUBJECT CODES"]
```

```python
df2["SUBJECT NAME"] = df["SUBJECT NAME"]
```

```python
df2["Course Nature"] = df["*COURSE NATURE (Hard/Soft/\n Workshop/ NTCC)"]
```

```python
df2
```

```python
holiday_df = pd.read_excel('holiday.xlsx')
```

```python
holiday_df
```

```python
print(holiday_df.columns)
```

```python
import numpy as np
start_date = '2024-02-18'
end_date = '2024-03-9'
date_range = pd.date_range(start=start_date, end=end_date)
weekdays_range = date_range[date_range.dayofweek < 5]
holiday_df['Date '] = pd.to_datetime(holiday_df['Date '], format='%d-%b - %Y')

# Iterate over the DataFrame and fill the 'Date' column
date_index = 0
for i, row in df2.iterrows():
    if pd.isna(row["SUBJECT NAME"]):  # Check if NaN value is encountered in the "SUBJECT NAME" column
        df2.at[i, 'Date'] = np.nan  # Fill NaN in the 'Date' column
        date_index = 0  # Reset the date index to start from the beginning of the range
    else:
        # Find the next valid weekday date that is not a holiday
        while True:
            if date_index >= len(weekdays_range):
                df2.at[i, 'Date'] = np.nan  # Fill NaN if no valid weekday date is found
                break
            elif weekdays_range[date_index].dayofweek < 5 and weekdays_range[date_index] not in holiday_df['Date '].values:
                df2.at[i, 'Date'] = weekdays_range[date_index]  # Fill the 'Date' column with the valid weekday date
                date_index += 1  # Move to the next date
                break
            else:
                date_index += 1  # Move to the next date
```

```python
df2
```

```python
# Define the file path for the Excel file
excel_file_path = 'output_data.xlsx'

# Write the DataFrame data to an Excel sheet with adjusted column width
with pd.ExcelWriter(excel_file_path, engine='xlsxwriter') as writer:
    df2.to_excel(writer, index=False)
    worksheet = writer.sheets['Sheet1']  # Adjust 'Sheet1' to the appropriate sheet name if needed
    worksheet.set_column('E:E', 15)  # Adjust column width for the 'Date' column (column 'E') as needed
```
