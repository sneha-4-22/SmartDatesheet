```python
import pandas as pd
import xlsxwriter
```

```python
excel_file = 'BBA & MBA PROGRAM STRUCTURE EVEN SEMESTER 2024 (2).xlsx'
```

```python

df = pd.read_excel(excel_file)
```

```python
df
```

```python
df2 = pd.DataFrame()
```

```python
df2["S.no."] = df["S No"]
```

```python
df2["SUBJECT CODES"] = df["SUBJECT CODES"]
```

```python
df2["SUBJECT NAME"] = df["SUBJECT NAME"]
```

```python
df2["Course TYPE"] = df["COURSE TYPE (Core/Elective /\n University Compulsory)"]
```

```python
df2
```

```python
holiday_df = pd.read_excel('holiday.xlsx')
```

```python
holiday_df
```

```python
print(holiday_df.columns)
```

```python
df2 = df2.drop(df2[df2['Course TYPE'] == 'PBL'].index)
```

```python
import numpy as np

start_date = '2024-03-01'
end_date = '2024-03-07'
gap_days = 0

date_range = pd.date_range(start=start_date, end=end_date, freq=f'{gap_days + 1}D')
weekdays_range = date_range[date_range.dayofweek < 6]
holiday_df['Date '] = pd.to_datetime(holiday_df['Date '], format='%d-%b - %Y')

date_index = 0
for i, row in df2.iterrows():
    if pd.isna(row["SUBJECT NAME"]):  
        df2.at[i, 'Date'] = np.nan  
        df2.at[i, 'Time Slot'] = '1:00 - 2:30'
        date_index = 0  
    elif not pd.isna(row["Course TYPE"]):
        while True:
            if date_index >= len(weekdays_range):
                df2.at[i, 'Date'] = np.nan  
                df2.at[i, 'Time Slot'] = np.nan
                break
            elif weekdays_range[date_index].dayofweek < 6 and weekdays_range[date_index] not in holiday_df['Date '].values:
                df2.at[i, 'Date'] = weekdays_range[date_index]  
                date_index += 1  
                break
            else:
                date_index += 1  

df2['Time Slot'] = np.where(pd.notna(df2['Course TYPE']), '9:00 - 10:30', np.nan)
df2.loc[df2['Date'].isna() & pd.notna(df2['Course TYPE']), 'Time Slot'] = '1:00 - 2:30'


start_date_index = 0
for i, row in df2.iterrows():
    if pd.isna(row["Date"]) and not pd.isna(row["Course TYPE"]):
        while True:
            if start_date_index >= len(weekdays_range):
                break
            elif weekdays_range[start_date_index].dayofweek < 6 and weekdays_range[start_date_index] not in holiday_df['Date '].values:
                df2.at[i, 'Date'] = weekdays_range[start_date_index]  
                start_date_index += 1  
                break
            else:
                start_date_index += 1

df2.loc[df2['Date'].isna() & pd.notna(df2['Course TYPE']), 'Time Slot'] = '1:00 - 2:30'

df2['Date'] = df2.groupby('SUBJECT NAME')['Date'].transform('first')
```

```python
df2['Date'] = pd.to_datetime(df2['Date'])
```

```python
df2
```

```python
excel_file_path = 'final.xlsx'

with pd.ExcelWriter(excel_file_path, engine='xlsxwriter') as writer:
    df2.to_excel(writer, index=False)
    worksheet = writer.sheets['Sheet1']  
    worksheet.set_column('E:E', 15) 
```
