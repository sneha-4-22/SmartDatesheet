```python
df1[['L', 'T', 'P']] = df1['L-T-P'].str.split('-', expand=True)
df1[['L', 'T', 'P']] = df1[['L', 'T', 'P']].apply(pd.to_numeric)
df1 = df1[(df1['L'] != 0)]
```

```python
df1['Year'] = ''

# Define conditions and assign values to the "Year" column based on the "Semester" values
df1.loc[df1['Semester'].isin(['I', 'II']), 'Year'] = 1
df1.loc[df1['Semester'].isin(['III', 'IV']), 'Year'] = 2
df1.loc[df1['Semester'].isin(['V', 'VI']), 'Year'] = 3
df1.loc[df1['Semester'].isin(['VII', 'VIII']), 'Year'] = 4
df1.loc[df1['Semester'].isin(['IX', 'X']), 'Year'] = 5
```

```python
filtered_psc_df = psc_df[psc_df['Sub. Name'].isin(course)].copy()

psc_stu = pd.DataFrame()
psc_stu["Student"] = filtered_psc_df["Student"]
psc_stu["Rollno"] = filtered_psc_df["Rollno"]
psc_stu["Sub. Name"] = filtered_psc_df["Sub. Name"]
psc_stu["Semester"] = filtered_psc_df["Semester"]
psc_stu["Class"] = filtered_psc_df["Class"]
psc_stu["STATUS"] = filtered_psc_df["STATUS"]

psc_stu = psc_stu[psc_stu['STATUS'] == 'REGISTERED']
```

```python
import datetime
c_year = int(str(datetime.datetime.now().year)[2:])
psc_stu['Current_year'] = c_year - psc_stu['Rollno'].str[2:4].fillna(0).astype(int)
```

```python
# Calculate the number of students for each subject in PSC data
psc_students_count = psc_stu["Sub. Name"].value_counts()

# Iterate through each subject in the PSC data
for subject, count in psc_students_count.items():
    # Check if the subject exists in df1
    if subject in df1['Course Name'].values:
        # Add the number of students from PSC data to the corresponding subject in df1
        df1.loc[df1['Course Name'] == subject, 'No. of Students'] += count
    else:
        print(f"Subject {subject} from PSC data not found in df1.")
```

```python
# Iterate through each row in the DataFrame containing student data
for index, row in psc_stu.iterrows():
    # Extract relevant information from the row
    subject = row['Sub. Name']
    current_year = row['Current_year']
    
    # Check if the subject exists in psc_df
    if subject in psc_df['Sub. Name'].values:
        # If the subject is found, add the current_year to the corresponding row in df1
        df1.loc[df1['Course Name'] == subject, 'psc_year'] = current_year
    else:
        # If the subject is not found, add {0} to indicate that the subject is not present in psc_df
        df1.loc[df1['Course Name'] == subject, 'psc_year'] = 0
```

```python
final_df = pd.DataFrame()
```

```python
final_df["Program"] = df1["Program"]
final_df["Semester"]= df1["Semester"]
final_df["Course_Name"]= df1["Course Name"]
final_df["Course_Code"]= df1["Course Code"]
final_df["Course_Type"]= df1["Course Type"]
final_df["Total students"]= df1["No. of Students"]
```

```python
import pandas as pd
import numpy as np

# Define start and end dates
start_date = '2024-03-11'
end_date = '2024-04-28'

# Create date range and select weekdays
date_range = pd.date_range(start=start_date, end=end_date)
weekdays_range = date_range[date_range.dayofweek < 5]

# Define time slots based on term
term = "end-term"  # or "mid-term"
if term == "mid-term":
    time_slots = [
        '9:00 - 10:30', 
        '11:30 - 13:00', 
        '14:00 - 15:30'
    ]
elif term == "end-term":
    time_slots = [
        '9:00 - 12:30', 
        '13:00 - 15:30'
    ]

# Define semester mapping
semester_mapping = {
    "I": 1,
    "II": 2,
    "III": 3,
    "IV": 4,
    # Add more mappings as needed
}

# Initialize variables
date_index = 0  
prev_program_semester = None  
last_time_slot = None  # Track the last assigned time slot
last_date = {}

# Iterate over rows in df1
for i, row in df1.iterrows():
    if pd.isna(row["Course Name"]):  
        final_df.at[i, 'Date'] = np.nan  
        date_index = 0  
    else:
        program_semester = (row["Program"], semester_mapping.get(row["Semester"], None))  # Convert semester name to integer
        
        if program_semester[1] is not None:
            if program_semester != prev_program_semester:
                date_index = 0  
                prev_program_semester = program_semester
                semester_index = program_semester[1] - 1  # Adjust time slot index based on semester
                time_slot_index = semester_index % len(time_slots)
                last_time_slot = None  # Reset last_time_slot when program/semester changes

            if program_semester in last_date:
                last_assigned_date = last_date[program_semester] + pd.Timedelta(days=2) 
            else:
                last_assigned_date = pd.Timestamp(weekdays_range[0])  

            while True:
                if date_index >= len(weekdays_range):
                    # Reset date index if all dates are used
                    date_index = 0

                if weekdays_range[date_index].dayofweek < 5:
                    current_date = pd.Timestamp(weekdays_range[date_index].date())  
                    if current_date >= last_assigned_date:
                        final_df.at[i, 'Date'] = current_date  
                        last_date[program_semester] = current_date 

                        # Assign the appropriate time slot based on semester
                        final_df.at[i, 'Time Slot'] = time_slots[time_slot_index]
                        last_time_slot = final_df.at[i, 'Time Slot']

                        # Increment time slot index for the next iteration
                        time_slot_index = (time_slot_index + 1) % len(time_slots)

                        date_index += 1  
                        break
                    else:
                        date_index += 1  
                else:
                    date_index += 1
```

```python
# Group the dataframe by 'Date' and 'Time Slot' and sum the 'Total students' column
total_students_count = final_df.groupby(['Date', 'Time Slot'])['Total students'].sum().reset_index()

# Iterate through each group
for date_slot, group_df in final_df.groupby(['Date', 'Time Slot']):
    date, time_slot = date_slot
    total_students = total_students_count.loc[(total_students_count['Date'] == date) & (total_students_count['Time Slot'] == time_slot), 'Total students'].iloc[0]
    if total_students > 800:
        # Assign a different time slot for subjects on the same date
        new_time_slot_index = time_slots.index(time_slot) + 1  # Increment time slot index
        if new_time_slot_index >= len(time_slots):
            new_time_slot_index = 0  # Wrap around to the first time slot if the last is reached
        new_time_slot = time_slots[new_time_slot_index]

        # Update the time slot for subjects on the same date and time slot
        final_df.loc[(final_df['Date'] == date) & (final_df['Time Slot'] == time_slot), 'Time Slot'] = new_time_slot
```
